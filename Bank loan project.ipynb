# Bank Loan Project Prediction

# Step1=Import Libararies

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

# Step2= load  dataset

```python
qq=pd.read_csv("C:\\Users\\Dell\\OneDrive\\Desktop\\jup\\loan project by multi class ,logectic\\train_1.csv")
qq
```

# Step 3: Data Analysis/Data Handling/Data Modififcation/Data Cleaning

```python
qq=qq.drop(columns=["job","marital","education","default","housing","contact"])
qq
```

#  Step 4: Seperate input and output in different variables


```python
from sklearn.preprocessing import LabelEncoder
ll = LabelEncoder()

qq["loan"]=ll.fit_transform(qq["loan"])
qq["month"]=ll.fit_transform(qq["month"])
qq["poutcome"]=ll.fit_transform(qq["poutcome"])
qq["subscribed"]=ll.fit_transform(qq["subscribed"])
qq
```

```python
x=qq.drop(columns=["loan"])
x
```

```python
y= qq.iloc[:,3]
y
```

# Step 5:Split input output data in training and testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2,random_state = 42)
```

# Import Algorithm

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

# Step 6: Model Train(.fit()) and Prediction(.predict())

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

# Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
